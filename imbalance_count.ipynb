```python
import sys
sys.path.append("../") # adding the path to read the utils file
import pandas as pd
from utils import *
```

```python
# you should run making_hikari2022_csv.ipynb before running this cell, since the two csv given with the paper have two differences
# to see what are the data types and the features of the dataset, you can take a look on the utils file
hikari = pd.concat([
    pd.read_csv('datasets/HIKARI-2021/ALLFLOWMETER_HIKARI2021.csv', usecols=['traffic_category']),
    pd.read_csv('datasets/HIKARI-2021/ALLFLOWMETER_HIKARI2022.csv', usecols=['traffic_category'])],
    ignore_index=True)
```

```python
counted = hikari.value_counts('traffic_category')

hikari_imb = pd.DataFrame({'label': counted.index, 'count': counted.values})
hikari_imb['%_over_majority_class'] = (hikari_imb['count'] / hikari_imb['count'].max()) * 100
hikari_imb['%_over_all'] = (hikari_imb['count'] / hikari_imb['count'].sum()) * 100
```

```python
hikari_imb.head(20)
```

```python
with pd.ExcelWriter(
    "documents/imb_analysis.xlsx",
    mode="a",
    engine="openpyxl",
    if_sheet_exists="replace",
) as writer:
    hikari_imb.to_excel(writer, sheet_name='Hikari-2021', index=False)
```
