# Preprocessing the data

```python
# Define the pattern to match the file names
file_pattern = '../datasets/CICIDS-2017/*.csv'

# Retrieve the list of file paths matching the pattern
file_paths = glob.glob(file_pattern)

def load_ds(path):
    print(path)
    return pd.read_csv(path, encoding="us-ascii", encoding_errors="ignore", dtype=dtype_cicids).drop(columns=[" Fwd Header Length.1"])

ds = pd.concat([load_ds(file_path) for file_path in file_paths], ignore_index=True)
ds.columns = ds.columns.str.strip()
ds.head()
```
