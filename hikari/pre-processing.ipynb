```python
import sys
sys.path.append("../") # adding the path to read the utils file
import pandas as pd
from utils import *
import numpy as np

pd.set_option('display.max_columns', None)
```

```python
# you should run making_hikari2022_csv.ipynb before running this cell, since the two csv given with the paper have two differences
# to see what are the data types and the features of the dataset, you can take a look on the utils file
ds = pd.concat([
    pd.read_csv('../datasets/HIKARI-2021/ALLFLOWMETER_HIKARI2021.csv', dtype=dtype_hikari, usecols=selected_features_hikari),
    pd.read_csv('../datasets/HIKARI-2021/ALLFLOWMETER_HIKARI2022.csv', dtype=dtype_hikari, usecols=selected_features_hikari)],
    ignore_index=True)
ds.head()
```

```python
ds.info()
```

# First thing we need to remove all the flows with a duration over 1800 seconds, with more than 2 FIN flags counted and with more than 1 RST flags counted.

```python
ds = ds[ds['flow_FIN_flag_count'] < 3]
ds = ds[ds['flow_RST_flag_count'] < 2]
ds = ds[ds['flow_duration'] < 1801]
ds.info()
```

# Now we need to decrease to 1 all the FIN counts that are a value of 2, and to decrease all the ACK flag counts by 2.

```python
def fix_fin_and_ack(x):
    if x.flow_FIN_flag_count == 2:
        x.flow_FIN_flag_count = 1
        x.flow_ACK_flag_count = x.flow_ACK_flag_count - 2
    return x

ds = ds.apply(fix_fin_and_ack, axis=1)
```

As we can see the minimum for ACK flag is 0 and not negative after the decrease, so we can do it without any problem.
