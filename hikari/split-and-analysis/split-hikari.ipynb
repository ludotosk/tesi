```python
import pandas as pd
```

```python
ds = pd.read_csv('../../datasets/HIKARI-2021/ALLFLOWMETER_HIKARI2021.csv', index_col='Unnamed: 0')
ds.head()
```

```python
# Create an empty list to store individual DataFrames
individual_dfs = []

# Initialize variables to keep track of the start and end indices for each DataFrame
start_index = 0

for i in range(len(ds)):
    if ds.index.get_level_values(0)[i] == 0:
        # If the value in the specified index column is 0, create a new DataFrame
        individual_dfs.append(ds.iloc[start_index:i])
        start_index = i + 1

# Append the last DataFrame
individual_dfs.append(ds.iloc[start_index:])
```

```python
len(individual_dfs)
```

```python
for i in range(1,6):
    print(individual_dfs[i].shape)
```

```python
for i in range(1,6):
    file_name = 'ds_' + str(i) + '.csv'
    individual_dfs[i].to_csv(file_name)
```
