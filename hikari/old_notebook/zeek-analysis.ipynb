# analisi bruteforce

```python
file1 = pd.read_csv('HIKARI-CICFlowMeter/zeek/conn-2021-04-11-Sunday.log',header=7,sep='\x09')
file1.head(10)
```

```python
file2 = pd.read_csv('HIKARI-CICFlowMeter/zeek/conn-2021-04-12-Monday.log',header=7,sep='\x09')
file2.head(10)
```

```python
# loading the connection that has perfomed the attack
bf_id = pd.read_csv('bf_id.csv', index_col=0)
bf_id
```

```python
file1['con_id'] = file1['addr.1'] + '-' + file1['addr'] + '-' + file1['port.1'].astype('str') + '-' + file1['port'].astype('str')
file1['con_id']
```

```python
file2['con_id'] = file2['addr.1'] + '-' + file2['addr'] + '-' + file2['port.1'].astype('str') + '-' + file2['port'].astype('str')
file2['con_id']
```

```python
# creating a function to label the connection
con_id_list = bf_id['con_id'].to_list()

def label_connection(flow_id):
    if flow_id in con_id_list:
        return 'Bruteforce'
    else:
        return 'Benign'

file1['Label'] = file1['con_id'].apply(label_connection)
file2['Label'] = file2['con_id'].apply(label_connection)
```
