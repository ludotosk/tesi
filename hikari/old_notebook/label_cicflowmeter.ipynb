```python
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
```

```python
# Function to remove the latest number after the hyphen
def remove_latest_number(text):
    x = text.split('-')
    return x[0] + '-' + x[1] + '-' + x[2] + '-' + x[3]
```

```python
def fix_address(ds):
    ds[' Destination IP'] = ds['Flow ID'].apply(lambda x: x.split('-')[0])
    ds[' Source IP'] = ds['Flow ID'].apply(lambda x: x.split('-')[1])
    ds[' Destination Port'] = ds['Flow ID'].apply(lambda x: int(x.split('-')[2]))
    ds[' Source Port'] = ds['Flow ID'].apply(lambda x: int(x.split('-')[3]))
```

# working on bruteforce xml attack

```python
# loading the connection that has perfomed the attack
xml_id = pd.read_csv('datasets/xml_id.csv', index_col=0)
xml_id
```

```python
# loading the two csv files where the xml bruteforce was performed
file1 = pd.read_csv('datasets/HIKARI-CICFlowMeter/Friday_2021-04-16_2304.pcap_ISCX.csv')
file2 = pd.read_csv('datasets/HIKARI-CICFlowMeter/Saturday_2021-04-17_0357.pcap_ISCX.csv')
```

```python
file1
```

```python
file2
```

```python
fix_address(file1)
fix_address(file2)
```

```python
# Applying the function to the DataFrame
file1['Flow ID'] = file1['Flow ID'].apply(remove_latest_number)
file2['Flow ID'] = file2['Flow ID'].apply(remove_latest_number)
```

```python
# creating a function to label the connection
con_id_list = xml_id['con_id'].to_list()

def label_connection(flow_id):
    if flow_id in con_id_list:
        return 'Bruteforce-XML'
    else:
        return 'Benign'

file1[' Label'] = file1['Flow ID'].apply(label_connection)
file2[' Label'] = file2['Flow ID'].apply(label_connection)
```
